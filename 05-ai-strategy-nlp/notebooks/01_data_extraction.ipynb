# 01 Data Extraction
Convert PDFs to cleaned text ready for NLP. Keep raw files immutable; write outputs to data/processed.

```python

from pathlib import Path
import sys
sys.path.append(str((Path('..')).resolve()))
from src import text_processing as tp

DATA_DIR = Path('..') / 'data'
RAW_DIR = DATA_DIR / 'raw'
PROCESSED_DIR = DATA_DIR / 'processed'
MODEL = 'en_core_web_sm'  # adjust if you downloaded a different spaCy model
```

```python
pdfs = list(RAW_DIR.glob('*.pdf'))
pdfs
```

```python

for pdf in pdfs:
    out_txt = PROCESSED_DIR / f"{pdf.stem}_clean.txt"
    tp.pipeline_clean_and_save(pdf_path=pdf, out_path=out_txt, model=MODEL)
    print(f'Saved {out_txt}')
```
